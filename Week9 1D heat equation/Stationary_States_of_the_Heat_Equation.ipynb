```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
L = 30  # Length of the interval
N = 51   # Number of sections
dx = L/N # Step size

# Initialize arrays
T = np.zeros(N+1)  # Array to store temperatures
T_new = np.zeros(N+1)  # Array to store updated temperatures
x = np.linspace(0, L, N+1)  # Array to store x values

# Set boundary conditions
T[0] = 100  # T(x=0) = 100
T[N] = 0    # T(x=L) = 0

#initial state:
plt.clf()
plt.plot(x,T, 'ro-',mfc='w')
plt.xlabel('x')
plt.ylabel('T(x)')
plt.title('initial state')
plt.show()

# Solve 1D heat equation using Jacobi method
max_iteration = 600
iteration = 0

while iteration < max_iteration:
    T_new[0] = 100
    T_new[N] = 0
    for i in range(1, N):
        T_new[i] = 0.5 * (T[i-1] + T[i+1])
    T = np.copy(T_new)
    iteration += 1

    if iteration % 200 == 0:
        plt.clf()
        plt.plot(x,T, 'ro-',mfc='w')
        plt.title('Temperation distribution with iteration = ' + str(iteration))
        plt.show()
        plt.draw()
        plt.pause(0.05)
```
