Compute and print the variance and the standard deviation sigma between your data set and the solution as
described on the lecture notes.

```python
def calc_var_std(d):
    mean_d = sum(d) / nNumber
    varience = 0
    for i in range(nNumber):
        varience += (d[i] - mean_d)**2
    varience = varience / nNumber
    standard_deviation = np.sqrt(varience)
    return varience, standard_deviation
```

```python
data_var = np.var(data)
data_std = np.std(data)
solution_var = np.var(solution)
solution_std = np.std(solution)

print('data_var = ',data_var)
print('data_std = ',data_std)
print('sol_var = ',solution_var)
print('sol_std = ',solution_std)

my_data_var, my_data_std = calc_var_std(data)
my_solution_var, my_solution_std = calc_var_std(solution)
print('my_data_var = ',my_data_var)
print('my_data_std = ',my_data_std)
print('my_sol_var = ',my_solution_var)
print('my_sol_std = ',my_solution_std)
```

Let us assume we know the average is zero. Let us recalculate the variance and the standard deviation.

```python
def calc_var_std_with_0_mean(d):
    mean_d = 0
    varience = 0
    for i in range(nNumber):
        varience += (d[i] - mean_d)**2
    varience = varience / nNumber
    standard_deviation = np.sqrt(varience)
    return varience, standard_deviation
```

```python
my_data_var, my_data_std = calc_var_std_with_0_mean(data)
my_solution_var, my_solution_std = calc_var_std_with_0_mean(solution)
print('my_data_var = ',my_data_var)
print('my_data_std = ',my_data_std)
print('my_sol_var = ',my_solution_var)
print('my_sol_std = ',my_solution_std)
```
